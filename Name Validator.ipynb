```python
import pandas as pd
import requests
from io import StringIO

calendar_to_replace = "Input\Calendars\Calendar_Super_League_25_26.csv"
url_football_data = "https://www.football-data.co.uk/mmz4281/2526/GR1.csv"

calendar = pd.read_csv(calendar_to_replace)

def fetch_league_data(url: str) -> pd.DataFrame:
    response = requests.get(url, timeout=30)
    response.raise_for_status()
    return pd.read_csv(StringIO(response.text))

data_downloaded = fetch_league_data(url_football_data)

calendar_names = set(calendar['Awayteam'].unique())
data_names = set(data_downloaded['AwayTeam'].unique())

missing_teams = data_names - calendar_names
missing_teams2 =  calendar_names - data_names

print("Please map:")
print(missing_teams)
print('to:')
print(missing_teams2)
```

```python
### The Key is the data_downloaded name, the Value is the calendar name
dict_replace = {
    "Volos NFC": "Volos",
    "Asteras Tripolis": "Asteras T.",
    "Levadeiakos": "Levadiakos",
    "Larisa": "AEL Larissa",
}



dict_replace = {v: k for k, v in dict_replace.items()}

calendar['Awayteam'] = calendar['Awayteam'].replace(dict_replace)
calendar['Hometeam'] = calendar['Hometeam'].replace(dict_replace)

calendar.to_csv(calendar_to_replace, index=False)
```
